```python
import matplotlib.pyplot as plt
```

```python
EVENT, TIME, FROM_NODE, TO_NODE, PACKET_TYPE, PACKET_SIZE, FLAGS, FID, SOURCE_ADDRESS, DST_ADDRESS, SEQ_NUM, PKT_ID = [i for i in range(12)]
```

```python
with open("out.tr", "r") as f:
    data = f.readlines()
```

```python
output, delay, mean, jitter, count, sum, rtime = ["Seq | Send_t | Rec_t | Delay | Mean_Delay | Jitter\n"], [], [], [], 0, 0, []
startTime = {}
for line in data:
    item = line.split()
    if item[EVENT] == '+' and item[FROM_NODE] == '0' and item[PKT_ID] not in startTime.keys():
        startTime[item[PKT_ID]] = item[TIME]
    if item[EVENT] == 'r' and item[TO_NODE] == '5':
        delay.append(float(item[TIME]) - float(startTime[item[PKT_ID]]))
        count += 1
        rtime.append(float(item[TIME]))
        sum += delay[-1]
        mean.append(sum / count)
        jitter.append(delay[-1] - mean[-1])
        output.append(" ".join([item[PKT_ID], startTime[item[PKT_ID]], item[TIME], str(delay[-1]), str(mean[-1]), str(jitter[-1])]) + "\n")
```

```python
fig, ax = plt.subplots(figsize=(24, 10), dpi=100)
plt.title("TCP connection n0-n5, Delay, Mean-Delay and Jitter vs Simulation Time")
plt.plot(rtime, delay, 'r+', label='Delay')
plt.plot(rtime, mean, 'b+', label="Mean delay")
plt.plot(rtime, jitter, 'go', label="Jitter")
plt.xlabel('Time')
plt.ylabel('Delay')
plt.legend()
plt.show()
```

```python
with open("delay-tcp0.txt", "w") as f:
    f.writelines(output)
```
